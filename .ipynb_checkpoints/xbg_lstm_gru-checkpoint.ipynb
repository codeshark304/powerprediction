```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from scipy.signal import savgol_filter
from sklearn.metrics import mean_absolute_error

# Load dataset
df = pd.read_csv("processed_weather.csv", parse_dates=["Date"], index_col="Date")  # Ensure date column exists
df = df.sort_index()  # Sort data by date if not already

# Select features and target
features = ["temp", "humidity", "wind",'barometer','Weather_Label']  # Adjust based on your dataset
target = "Elec_kW"

# Normalize data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[features + [target]])

# Create sequences for LSTM
def create_sequences(data, seq_length=30):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, :-1])  # Features
        y.append(data[i+seq_length, -1])  # Target (Elec_kW)
    return np.array(X), np.array(y)

SEQ_LENGTH = 30  # Use 30 days of past data for prediction
X, y = create_sequences(scaled_data, SEQ_LENGTH)

# Train-test split
split = int(0.8 * len(X))
X_train, X_test, y_train, y_test = X[:split], X[split:], y[:split], y[split:]

# Build LSTM Model
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(SEQ_LENGTH, X.shape[2])),
    Dropout(0.3),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Make Predictions
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.4f}")

# Inverse transform predictions
scaler_target = MinMaxScaler()
scaler_target.fit(df[[target]])  # Fit only on target column
y_test_inv = scaler_target.inverse_transform(y_test.reshape(-1, 1))
y_pred_inv = scaler_target.inverse_transform(y_pred)

# Apply Savitzky-Golay filter for smoothing
window = 15  # Adjust window size for smoothness
poly_order = 3

y_test_smooth = savgol_filter(y_test_inv.flatten(), window, poly_order)
y_pred_smooth = savgol_filter(y_pred_inv.flatten(), window, poly_order)

# Compute confidence interval
std_dev = np.std(y_pred_inv.flatten() - y_test_inv.flatten())
upper_bound = y_pred_smooth + 1.96 * std_dev
lower_bound = y_pred_smooth - 1.96 * std_dev

# Plot Actual vs Predicted Graph
plt.figure(figsize=(12, 6))
dates_test = df.index[split + SEQ_LENGTH:]  # Adjust index range

plt.plot(dates_test, y_test_smooth, label="Actual Elec_kW (Smoothed)", color="blue")
plt.plot(dates_test, y_pred_smooth, label="Predicted Elec_kW (Smoothed)", linestyle="dashed", color="red")
plt.fill_between(dates_test, lower_bound, upper_bound, color="red", alpha=0.2, label="95% Confidence Interval")

plt.xlabel("Date")
plt.ylabel("Electricity Consumption (kW)")
plt.title("Actual vs Predicted Electricity Consumption")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from scipy.signal import savgol_filter
from sklearn.metrics import mean_absolute_error

# Load dataset
df = pd.read_csv("processed_weather.csv", parse_dates=["Date"], index_col="Date")  # Ensure date column exists
df = df.sort_index()  # Sort data by date if not already

# Select features and target
features = ["temp", "humidity", "wind", "barometer", "Weather_Label"]  # Adjust based on your dataset
target = "Elec_kW"

# Normalize data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[features + [target]])

X, y = [], []
for i in range(len(scaled_data) - 1):  # Use all past data, predict next
    X.append(scaled_data[i, :-1])  # Use only features
    y.append(scaled_data[i + 1, -1])  # Predict next Elec_kW

X, y = np.array(X), np.array(y)

# Split data
split = int(0.8 * len(X))
X_train, X_test, y_train, y_test = X[:split], X[split:], y[:split], y[split:]

# Reshape for LSTM
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Build Model
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(1, X.shape[1])),
    Dropout(0.3),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Make Predictions
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.4f}")

# Inverse transform predictions
scaler_target = MinMaxScaler()
scaler_target.fit(df[[target]])  # Fit only on target column
y_test_inv = scaler_target.inverse_transform(y_test.reshape(-1, 1))
y_pred_inv = scaler_target.inverse_transform(y_pred)

# Apply Savitzky-Golay filter for smoothing
window = 15  # Adjust window size for smoothness
poly_order = 3

y_test_smooth = savgol_filter(y_test_inv.flatten(), window, poly_order)
y_pred_smooth = savgol_filter(y_pred_inv.flatten(), window, poly_order)

# Compute confidence interval
std_dev = np.std(y_pred_inv.flatten() - y_test_inv.flatten())
upper_bound = y_pred_smooth + 1.96 * std_dev
lower_bound = y_pred_smooth - 1.96 * std_dev

# Ensure `dates_test` has the same length as `y_test`
dates_test = df.index[split + 1:]  # Shift index by 1 to match prediction alignment

# Plot Actual vs Predicted Graph
plt.figure(figsize=(12, 6))
plt.plot(dates_test, y_test_smooth, label="Actual Elec_kW (Smoothed)", color="blue")
plt.plot(dates_test, y_pred_smooth, label="Predicted Elec_kW (Smoothed)", linestyle="dashed", color="red")
plt.fill_between(dates_test, lower_bound, upper_bound, color="red", alpha=0.2, label="95% Confidence Interval")

plt.xlabel("Date")
plt.ylabel("Electricity Consumption (kW)")
plt.title("Actual vs Predicted Electricity Consumption")
plt.legend()
plt.grid()
plt.show()
```
